# lets see what SweetViz can do with just this state of input data.

```python
crime_df=crime_df.drop(columns=["Date"],axis=1)
my_report1=sweetviz.analyze([crime_df,"crimeData"], "Arrest")
```

```python
my_report1.show_html("Report on Analyze.html")
```

# Comparing two dataframes (e.g. Test vs Training sets)

```python
X=crime_df.drop(columns=["Arrest"],axis=1)
X.head()
```

```python
y=crime_df["Arrest"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, shuffle=True,
                                                        train_size=0.8, random_state=6
                                                        )
```

```python
my_report=sweetviz.compare([x_train,"Train"],[x_test,"Test"], "Domestic")
```

```python
my_report.show_html("Report.html")
```
