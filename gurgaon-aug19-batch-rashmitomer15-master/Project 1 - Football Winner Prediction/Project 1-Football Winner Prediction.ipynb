# 1.	Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)

```python
import numpy as np
import pandas as pd

laliga= pd.read_csv("Laliga.csv")
print(laliga.shape)
laliga.head()
```

```python
new_header = laliga.iloc[0] #grab the first row for the header
laliga = laliga[1:] #take the data less the header row
laliga.columns = new_header #set the header row as the df header
laliga.head()
```

```python
laliga.set_index("Pos")
laliga.head()
```

```python
laliga[laliga[:]=='-'].count()
```

```python
laliga[laliga[:]=='-'] = 0
laliga[laliga[:]=='-'].count()
```

# 2.	Print all the teams which have started playing between 1930-1980. (5 points)

```python
laliga['Debut'].head()
```

```python
laliga['Debut']=laliga['Debut'].str.extract(r'^(\d{4})',expand=False)
laliga['Debut'].head()
```

```python
print('laliga:',laliga['Debut'].shape)
teamsDebut1930And80=laliga[(pd.to_datetime(laliga['Debut']) > '1930') & (pd.to_datetime(laliga['Debut']) < '1980')]
print("Teams that started between 1930 And 1980 count:",teamsDebut1930And80.shape)
print("Team ,  Debut year")
print(teamsDebut1930And80['Team'] +", "+teamsDebut1930And80['Debut'] )
```

# 3.	Print the list of teams which came Top 5 in terms of points (2.5 points)

```python

#laliga[laliga[:]=='-'].count()
laliga['Points']=pd.to_numeric(laliga['Points'])
TeamWithTop5Points=laliga.sort_values(by='Points',ascending=False)
TeamWithTop5Points[['Team','Points']].head()
```

# 4.	Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(GoalsFor, GoalsAgainst):
    Goal_diff_count=GoalsFor - GoalsAgainst
    return Goal_diff_count
    
```

```python
laliga['Goal_diff_count']=Goal_diff_count(pd.to_numeric(laliga['GoalsFor']),pd.to_numeric(laliga['GoalsAgainst']))

MaxGoalDiffTeam=laliga.sort_values(by='Goal_diff_count',ascending=False)
print(MaxGoalDiffTeam[['Team','GoalsFor','GoalsAgainst','Goal_diff_count']].head(1))

MinGoalDiffTeam=laliga.sort_values(by='Goal_diff_count')
print('\n',MinGoalDiffTeam[['Team','GoalsFor','GoalsAgainst','Goal_diff_count']].head(1))
```

```python
laliga.drop('Goal_diff_count',axis=1,inplace=True)
laliga.shape
```
